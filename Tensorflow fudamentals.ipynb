Part 1: Theoretical Questions

The different data structures used in TensorFlow are:

Tensors: Tensors are the primary data structure in TensorFlow. They are multi-dimensional arrays that can hold numerical data. Examples include scalars, vectors, and matrices.

Constants: Constants are tensors whose values cannot be changed once assigned. They are typically used to store fixed values in a TensorFlow graph.
Variables: Variables are tensors whose values can be changed during the execution of a TensorFlow graph. They are commonly used to store and update the parameters of a model during training.

Placeholders: Placeholders are tensors that act as input nodes in a TensorFlow graph. They are used to feed data into the graph during the execution.
The TensorFlow constant differs from a TensorFlow variable in the following ways:
Constants have fixed values that cannot be changed, while variables can be updated during the execution of a TensorFlow graph.
Constants are typically used to store fixed values, such as hyperparameters or input data, while variables are used to store and update the parameters of a model.
Constants are initialized with a specific value and retain that value throughout the execution, while variables are initialized with an initial value but can be modified using operations like assignment.

Matrix addition, multiplication, and element-wise operations in TensorFlow can be performed using various functions and operators:
Matrix Addition: Use the tf.add() function or the + operator to add two matrices element-wise.

Matrix Multiplication: Use the tf.matmul() function or the @ operator to perform matrix multiplication between two matrices.

Element-wise Operations: Use arithmetic operators (+, -, *, /) or specific functions like tf.add(), tf.subtract(), tf.multiply(), tf.divide() to perform element-wise operations on matrices.
